####2.
Bu datadaki [9-14] sütunlarını birbirlerine karşı scatter plot ile çizerek aralarında herhangi bir correlation(positive, negative, none) olup olmadığını belirleyiniz. (fig, axs = plt.subplots(5,3) kullanabilirsiniz) 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

url='https://drive.google.com/file/d/18gyHbx6rfogq3yQ-GR9COjcGgyYlCnBZ/view?usp=sharing'
url2='https://drive.google.com/uc?id=' + url.split('/')[-2]
df = pd.read_csv(url2)
#sütunlara diğerleriyle ulaşamadığım için kolon isimlerinden oluşan list oluşturdum bununla erişiyorum
columnName = ['retail_and_recreation_percent_change_from_baseline','grocery_and_pharmacy_percent_change_from_baseline',
              'parks_percent_change_from_baseline','transit_stations_percent_change_from_baseline',
              'workplaces_percent_change_from_baseline','residential_percent_change_from_baseline']
#her bir renk ilk değeri temsil ediyor y=9. kolon, r=10. kolon vs daha anlaşılır olması için
colour = ['y','r','b','c','m']
 

j = 1 #2. kolon
l = 0 #renk kontrolü için
#else ekledim çünkü en altta j'yi i+2 olarak arttırıyorum bu da bellek hatası oluşturuyor
for i in range(0, 6, 1):
    while j <= 5:
        fig, ax = plt.subplots(1)
        
        if(i < 5):
          ax.scatter(df[columnName[i]],df[columnName[j]],c=colour[l])
            
        else:
          ax.scatter(df[columnName[i-1]],df[columnName[i]],c=colour[l])
          
        j += 1
    j = i + 2
    l += 1
```
